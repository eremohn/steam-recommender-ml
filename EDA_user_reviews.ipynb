### Visualización de datos:
#### se utilizaran gráficos para visualizar la distribución de los datos y identificar posibles patrones o anomalías.

#### Histograma:  visualiza la distribución de una variable numérica. Pueden ayudarnos a identificar si los datos siguen una distribución específica.

```python
# Ejemplo: Histograma de la variable 'funny'
plt.hist(df['funny'])
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Histograma de la variable "funny"')
plt.show()
```

#### Gráfico de barras: comparar la frecuencia de valores en una variable categórica o para comparar diferentes grupos en una variable numérica.

```python
# Obtener los conteos de análisis de sentimiento
count_sentiment = df['sentiment_analysis'].value_counts()

# Crear el gráfico de barras
plt.bar(count_sentiment.index, count_sentiment.values)

# Etiquetas y título
plt.xlabel('Análisis de Sentimiento')
plt.ylabel('Frecuencia')
plt.title('Gráfico de barras de Análisis de Sentimiento')

# Mostrar el gráfico
plt.show()
```
